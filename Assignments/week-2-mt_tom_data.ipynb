## Fix the Time

* The time stamp for the data is awkward. It may be more useful to convert the data in the hour column. To do this, we'll create a new column in our dataframe and label it Day.  Then we'll calculate the day of the month by dividing our hour data by the number of hours in the day. The result you get should be in decimal days. 

````python
# you can add a new column derived from existing columns 
df['Day'] = df['Hour']/24.0
````

The block of code above does a few things.
1. It adds a new column called Day, which is as easy as typing a new column name in your dataframe brackets and setting it equal to something.
2. It does some math, and the great part about dataframes is that when you perform a math operation on a column, it knows to apply the operations to each row in the column individually.

````python
# Let's check our dataframe and see what the new column and values look like
df.head(5)
````

We can see in the updated dataframe above that we have a new column called Day. Our next goal is to plot the wind speed (the m/s column) against the Day column instead of the default index, which is the sample column.

When we call the `.plot()` function on our dataframe, we can pass in parameters that tell the plot function what column to use for the X-axis and what column to use for the Y-axis. 

````python
# Here, instead of using the df['column name'].plot() we provide specific columns for X and Y axes.

df.plot(x="Day", y="m/s")
````
