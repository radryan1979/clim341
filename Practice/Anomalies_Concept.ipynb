# Climate Anomalies — What They Are and Why We Use Them

---

## Start with a simple question

Imagine you check the weather and see that the temperature today is **18°C (64°F)**.

Is that warm? Cold? Normal?

It depends entirely on **where you are** and **what time of year it is**.

- In northern Alaska in January, 18°C would be shockingly warm.
- In Phoenix, Arizona in July, 18°C would be unusually cold.
- In San Francisco in October, 18°C is perfectly ordinary.

The raw number alone doesn't tell you much. What you really want to know is:

> **How does today's temperature compare to what's *normal* for this place and time of year?**

That's exactly what a **climate anomaly** tells you.

---
## What is an anomaly?

A **climate anomaly** is the difference between an observed value and the long-term average for that same location and time of year.

$$\text{Anomaly} = \text{Observed Value} - \text{Long-Term Average}$$

| If the anomaly is... | It means... |
|---|---|
| **Positive (+)** | Warmer (or wetter, or windier) than usual |
| **Zero (0)** | Right on average — nothing unusual |
| **Negative (−)** | Cooler (or drier, or calmer) than usual |

### A concrete example

Suppose the average July temperature in Boston is **24°C**, based on decades of measurements.

- This July it reached **28°C** → anomaly = 28 − 24 = **+4°C** (warmer than average)
- Last July it was **22°C** → anomaly = 22 − 24 = **−2°C** (cooler than average)
- The July before that was **24°C** → anomaly = 24 − 24 = **0°C** (right on average)

---
## What is the "long-term average"?

The long-term average — also called the **climatology** or **baseline** — is computed by averaging measurements over many years, typically **30 years**.

The World Meteorological Organization (WMO) currently uses **1991–2020** as the standard 30-year baseline. You may also see 1981–2010 or 1961–1990 used in older datasets.

The baseline represents our best estimate of what is "normal" for a given place and time of year.

> **Important:** An anomaly is always defined *relative to its baseline*. If you change the baseline period, the anomaly values change too. This is why scientists always state which baseline they are using.

---
## Why do climate scientists use anomalies instead of raw temperatures?

This is a great question. If we already have temperature measurements, why go to the trouble of computing anomalies?

There are three main reasons:

### 1. Anomalies remove the seasonal cycle

Raw temperatures go up in summer and down in winter — that's the seasonal cycle. This makes it hard to see other patterns, like a long-term warming trend or an unusually warm month. Anomalies subtract out the seasonal cycle, so you're only looking at *departures from normal*.

### 2. Anomalies allow fair comparison between locations

Suppose one January, both Alaska and Florida are 5°C warmer than their respective averages. Their raw temperatures look totally different (Alaska might be −15°C; Florida might be +18°C), but their anomalies are identical (+5°C). Anomalies put every location on the same footing.

### 3. Anomalies highlight what's unusual

A large positive anomaly flags a genuine extreme event — a heat wave, a drought, an unusually warm season. This stands out immediately in an anomaly plot in a way that raw temperatures do not.

This is why virtually every major climate dataset — from NASA GISS to NOAA to the IPCC — reports anomalies rather than (or in addition to) raw values.

## Visualizing anomalies: the red-blue bar chart

The most common way to display temperature anomalies is a bar chart where:

- 🔴 **Red bars** = positive anomaly (warmer than average)
- 🔵 **Blue bars** = negative anomaly (cooler than average)

This is one of the most recognized visuals in climate science. When you look at a multi-decade record, a warming trend shows up not as any single bar getting taller, but as a gradual **shift from mostly blue bars on the left to mostly red bars on the right**.

Here is a schematic example of what this looks like:

```
Anomaly (°C)
  +2 |         ██            ████ ██
  +1 |    ██  ████    ██   ████████████
   0 |------------------------------------
  -1 | ████████    ████████
  -2 | ██              ██
     +------------------------------------→ Year
       1980                         2020
       (mostly blue)         (mostly red)
```

## Summary

| Term | Definition |
|---|---|
| **Anomaly** | Observed value minus the long-term average |
| **Baseline / Climatology** | The long-term average, typically computed over 30 years |
| **Positive anomaly** | Above average — warmer, wetter, etc. |
| **Negative anomaly** | Below average — cooler, drier, etc. |

**Key takeaway:** An anomaly doesn't tell you whether a place is hot or cold — it tells you whether it is *unusually* hot or cold *relative to its own normal*. That's what makes anomalies so useful for comparing places, tracking trends, and identifying extremes.
