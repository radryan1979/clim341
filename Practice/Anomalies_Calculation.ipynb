# Calculating and Plotting a Temperature Anomaly

In the previous notebook, we learned what an anomaly is. Now let's calculate one — step by step — using a small, simple dataset.

---

## Our dataset

We have **10 years** of July average temperatures (°C) recorded at a weather station in a fictional town called Millbrook.

| Year | July Avg Temp (°C) |
|------|--------------------|
| 2014 | 22.1 |
| 2015 | 21.3 |
| 2016 | 23.8 |
| 2017 | 20.9 |
| 2018 | 24.2 |
| 2019 | 22.7 |
| 2020 | 25.1 |
| 2021 | 23.4 |
| 2022 | 26.0 |
| 2023 | 24.8 |

Let's enter this data in Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Our dataset ---
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]
temperatures = [22.1, 21.3, 23.8, 20.9, 24.2, 22.7, 25.1, 23.4, 26.0, 24.8]

print("Years:", years)
print("Temperatures (°C):", temperatures)
```

---
## Step 1: Calculate the baseline (long-term average)

The baseline is simply the **mean** of all our temperature values. This represents "normal" — what a typical July looks like at Millbrook over our 10-year record.

We use `np.mean()` to compute it.

```python
# Calculate the baseline: the average temperature across all years
baseline = np.mean(temperatures)

print(f"Baseline (long-term average): {baseline:.1f} °C")
```

---
## Step 2: Calculate the anomalies

Now we apply the formula:

$$\text{Anomaly} = \text{Observed Temperature} - \text{Baseline}$$

We subtract the baseline from **every** temperature value. NumPy makes this easy — we can subtract a single number from a whole list in one line.

```python
# Convert to a NumPy array so we can do math on the whole list at once
temperatures = np.array(temperatures)

# Calculate anomalies: observed temperature minus the baseline
anomalies = temperatures - baseline

# Print a neat summary
print(f"{'Year':>6}  {'Temp (°C)':>10}  {'Anomaly (°C)':>13}")
print("-" * 35)
for year, temp, anom in zip(years, temperatures, anomalies):
    sign = "+" if anom >= 0 else ""
    print(f"{year:>6}  {temp:>10.1f}  {sign}{anom:>12.1f}")
```

---
## Step 3: Plot the anomalies

Now we make the classic red/blue anomaly bar chart.

The key idea is to color each bar based on its sign:
- If the anomaly is **positive** (above zero) → **red**
- If the anomaly is **negative** (below zero) → **blue**

We do this by building a list of colors — one per bar — using a list comprehension.

```python
# Assign a color to each bar: red for positive anomaly, blue for negative
colors = ['red' if anom >= 0 else 'blue' for anom in anomalies]

# --- Create the plot ---
fig, ax = plt.subplots(figsize=(10, 5))

# Draw the bars
ax.bar(years, anomalies, color=colors, width=0.7)

# Draw a horizontal line at zero (the baseline)
ax.axhline(0, color='black', linewidth=1.0)

# Labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Temperature Anomaly (°C)')
ax.set_title('July Temperature Anomalies — Millbrook Weather Station')

# Add the baseline value as a note on the plot
ax.text(years[-1], ax.get_ylim()[0] + 0.05,
        f'Baseline: {baseline:.1f} °C (2014–2023 mean)',
        ha='right', va='bottom', fontsize=9, color='gray')

plt.tight_layout()
plt.show()
```

---
## What does the plot tell us?

Take a moment to look at the chart.

- The **blue bars** on the left (2014–2017) show that early Julys in our record were *cooler* than average.
- The **red bars** on the right (2018–2023) show that recent Julys have been *warmer* than average.
- This shift from blue to red suggests a **warming trend** over this short record.

Notice that the raw temperatures ranged from about 21°C to 26°C — that variation alone doesn't make the trend obvious. The anomaly plot makes the pattern much clearer.

## Summary

In three steps, we:

1. **Computed the baseline** — the mean temperature over our full record
2. **Subtracted the baseline** from each year's temperature to get the anomaly
3. **Plotted the anomalies** as a red/blue bar chart

This is exactly the approach used in real climate datasets — just applied to a much larger collection of measurements across the globe.
