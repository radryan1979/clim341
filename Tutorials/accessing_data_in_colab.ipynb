# 📂 Accessing Data Files in Google Colab

**Welcome!** This notebook walks you through three different ways to load a data file into Google Colab so you can work with it in Python.

You only need to use **one** of these methods — pick whichever one fits your situation.

---

### The three methods covered here:

| Method | Best when... |
|--------|-------------|
| **Method 1: Upload from your computer** | You have the file on your laptop/desktop |
| **Method 2: Load from Google Drive** | The file is already in your Google Drive |
| **Method 3: Download from a URL** | The file is hosted online (e.g., a course website) |

---

> 💡 **What is Google Colab?**  
> Google Colab is a free, cloud-based environment where you can write and run Python code directly in your browser — no installation needed. Think of it like a Google Doc, but for code.

## 🗂️ Background: Where Does Colab Store Files?

When you open a Colab notebook, Google gives you a **temporary workspace** — a small virtual computer running in the cloud. This workspace has its own file system (like a folder), but it is **not permanent**: any files you upload directly will disappear when your session ends.

That's why we have multiple strategies:
- Uploading temporarily (quick, but gone after session ends)
- Connecting to Google Drive (persistent, lives in your Drive)
- Downloading from a URL (re-downloads each session, but automatic)

You can always see what files are available in your current session by clicking the **folder icon** (📁) in the left sidebar.

---
## Method 1: Upload a File Directly from Your Computer

This is the simplest method. You pick a file from your computer and Colab uploads it to your temporary workspace.

**⚠️ Reminder:** This file will be gone when your Colab session ends. If you need it again, you'll have to re-upload it.

**Steps:**
1. Run the cell below
2. A "Choose Files" button will appear — click it
3. Select your data file from your computer
4. Wait for the upload to complete (you'll see a progress bar)

The cell will print the name of the file once the upload is finished.

```python
# Method 1: Upload a file from your computer
# Run this cell, then click the "Choose Files" button that appears below it

from google.colab import files  # This imports Colab's built-in file tools

uploaded = files.upload()  # This opens the file picker

# Print the names of uploaded files
for filename in uploaded.keys():
    print(f"✅ Successfully uploaded: {filename}")
```

```python
# After uploading, open the file using Python
# Replace 'your_filename.csv' with the actual name of your file

import pandas as pd  # pandas is a library for working with data tables

filename = 'your_filename.csv'   # ← CHANGE THIS to your file's name

df = pd.read_csv(filename)       # Read the file into a "dataframe" (a table)

print("File loaded! Here are the first 5 rows:")
df.head()                        # .head() shows the first 5 rows
```

> 📝 **Note on file types:**  
> The example above uses `pd.read_csv()` for a `.csv` file. If your instructor gives you a different file type, you'll use a different function (e.g., `pd.read_excel()` for `.xlsx`). Your instructor will let you know which one to use.

## Method 2: Load a File from Google Drive

If your instructor shared a file with you via Google Drive, or you've already saved your data file to your own Drive, you can connect Colab directly to your Drive.

**Steps:**
1. Make sure your data file is already in your Google Drive
2. Run the first cell below — a pop-up will ask you to sign in and grant permission
3. Click through the permission prompts (you'll need to allow Colab to access your Drive)
4. Once connected, your Drive appears as a folder at the path `/content/drive/MyDrive/`
5. Update the file path in the second cell to match where your file is located

```python
# Method 2, Step A: Connect ("mount") your Google Drive
# Run this cell and follow the authorization prompts

from google.colab import drive

drive.mount('/content/drive')  # This links your Google Drive to Colab

print("✅ Google Drive is connected!")
print("Your Drive files are now accessible under: /content/drive/MyDrive/")
```

```python
# Method 2, Step B: List files in your Drive to find the right path
# This shows what's in the top level of your Google Drive
# (Use the file browser on the left sidebar too — it's easier!)

import os

drive_path = '/content/drive/MyDrive/'

print("Files and folders in your Google Drive:")
for item in os.listdir(drive_path):
    print(" ", item)
```

```python
# Method 2, Step C: Open the file
# Update the path below to point to your file

import pandas as pd

# Example: if your file is in a folder called 'METR101' inside your Drive
# file_path = '/content/drive/MyDrive/METR101/your_filename.csv'

# Example: if your file is directly in your Drive (not in any folder)
# file_path = '/content/drive/MyDrive/your_filename.csv'

file_path = '/content/drive/MyDrive/your_filename.csv'   # ← CHANGE THIS

df = pd.read_csv(file_path)

print("File loaded! Here are the first 5 rows:")
df.head()
```

> 💡 **Tip: Finding the exact file path**  
> After mounting your Drive, click the **folder icon (📁)** in the left sidebar. Navigate to your file, then right-click on it and choose **"Copy path"**. Paste that path into the cell above!

## Method 3: Download a File from a URL

If your instructor has posted the data file on a course website, Canvas, or any public URL, you can have Colab download it automatically each time you run the notebook.

We'll use two approaches:
- **Option A: `wget`** — a simple command-line download tool built into Linux (Colab runs on Linux)
- **Option B: `gdown`** — a special tool for downloading files from Google Drive *shared links*

### Option A: Download from any public URL (wget)

```python
# Method 3, Option A: Download using wget
# Replace the URL below with the actual link to your file

url = 'https://example.com/path/to/your_file.csv'   # ← CHANGE THIS

# The exclamation mark (!) lets us run a system command inside Colab
# -q means "quiet" (less output), -O sets the output filename
!wget -q -O downloaded_file.csv "{url}"

print("✅ Download complete!")
```

```python
# Now open the downloaded file
import pandas as pd

df = pd.read_csv('downloaded_file.csv')

print("File loaded! Here are the first 5 rows:")
df.head()
```

### Option B: Download from a Google Drive shared link (gdown)

If your instructor shared a file with you via a **Google Drive sharing link** (the kind you get by right-clicking a file in Drive and choosing "Share"), use `gdown`.

A Google Drive sharing link looks like:  
`https://drive.google.com/file/d/XXXXXXXXXXXXXXXXX/view?usp=sharing`

You only need the **file ID** — the long string of letters and numbers between `/d/` and `/view`.

```python
# Install gdown (it may already be installed in Colab, but this ensures it's up to date)
!pip install -q gdown

print("✅ gdown is ready to use!")
```

```python
# Method 3, Option B: Download using gdown
import gdown

# Paste the full sharing link from Google Drive here:
sharing_link = 'https://drive.google.com/file/d/YOUR_FILE_ID_HERE/view?usp=sharing'  # ← CHANGE THIS

# gdown will extract the file ID automatically from the full link
gdown.download(sharing_link, 'downloaded_file.csv', fuzzy=True, quiet=False)

print("\n✅ Download complete!")
```

```python
# Open the downloaded file
import pandas as pd

df = pd.read_csv('downloaded_file.csv')

print("File loaded! Here are the first 5 rows:")
df.head()
```
